## You Already Use Templates

**Word templates:**
- Duplicate ‚Üí Edit text ‚Üí Done

**Excel templates:**
- Duplicate ‚Üí Change numbers ‚Üí Done

**Notebooks:**
- Duplicate ‚Üí Edit inputs ‚Üí Run ‚Üí Done

**Same concept. Different tool.**


## How It Works

### Step 1: Duplicate
Copy the notebook (File ‚Üí Make a copy)

### Step 2: Edit Inputs
Change the parts you want:
- Your audience
- Your topic
- Your tone
- Your platform

### Step 3: Run All
Click "Run All" ‚Üí Get your results

### Step 4: Save Output
Use the JSON output in your workflow

**That's it!** No coding from scratch.


## Example: Content Generator

**Original notebook:**
- Topic: "AI productivity"
- Output: 10 hooks for AI productivity

**You duplicate it:**
- Change topic to: "Fitness tips"
- Run it
- Get: 10 hooks for fitness tips

**Same notebook. Different input. Different output.**

This is how creators scale content.


## Try It Yourself

**Here's a simple template example:**


````python
# This is a simple template
# Change the inputs below and run again!

topic = "AI for creators"  # ‚Üê Change this
platform = "TikTok"        # ‚Üê Change this
tone = "energetic"         # ‚Üê Change this

# The "template" part (you don't change this)
output = f"""
Topic: {topic}
Platform: {platform}
Tone: {tone}

Ready to generate content!
"""

print(output)
print("")
print("See? Change inputs ‚Üí Get different output")
print("This is how templates work!")
````

