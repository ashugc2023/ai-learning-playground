## Think of It Like This

**Word Document** + **Excel Spreadsheet** + **Results** = **Jupyter Notebook**

A notebook has three parts:
1. **Text** (explanations like this)
2. **Code** (instructions for the computer)
3. **Results** (what the code produces)

All in one place. All visible. All editable.

## The Notebook Workflow

1. **Read** the text (like this explanation)
2. **Run** the code (click play or press Shift+Enter)
3. **See** results appear below
4. **Edit** and run again

**That's it!** No complicated setup. No hidden steps.
